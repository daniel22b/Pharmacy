## Rozkład wieku użytkowników

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

df = pd.read_csv('customers.csv')
print(type(df))
print(df.columns)
```

```python
counts = df['GENDER'].value_counts()
print(counts)
counts.plot(kind='bar',color='#26d3d4')
plt.xlabel('Płec')
plt.ylabel('Liczba osob')
plt.title('Rozklad plci')
plt.show()

```

```python
df['GENDER'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
age = list(df['AGE'])
mean = np.mean(age)
std = np.std(age)
print(mean)
print(std)

xs = np.linspace(mean - 4*std,mean+4*std,1000)
ys = norm.pdf(xs,mean,std)
plt.plot(xs, ys, label='Rozkład normalny')
plt.hist(age, bins=30, density=True, alpha=0.5, label='Histogram wieku')
plt.xlabel('Wiek')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title('Histogram wieku z nałożonym rozkładem normalnym')
plt.legend()
plt.show()
```
